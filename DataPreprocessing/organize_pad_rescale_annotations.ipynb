```python
import os, csv
from shutil import copy
from PIL import Image
```

```python
#specify desired image side length (in pixels)
img_size = 500
#main folder from downloaded dataset. For me, this folder is called KolektorSDD
path = r'C:\Users\hgree\OneDrive\Documents\CS230\proj_data\KolektorSDD'
#path to images
image_folder_path = r'C:\Users\hgree\OneDrive\Documents\CS230\proj_data\KolektorSDD\input_images'

csvfile = open('plastic.csv', 'w', newline='')
writer = csv.DictWriter(csvfile, fieldnames = ['image name', 'defect status'])
writer.writeheader()

img_num = 0
for option in os.listdir(path):
    if('kos' not in option):
        continue
    else:
        subpath = os.path.join(path, option)
        print(subpath)
        imgs = os.listdir(subpath)
        for subimg in imgs:
            if("label" not in subimg):
                img_num += 1
                
                ##copy the image into the 'input_images' folder, rename
                #path_to_img = os.path.join(subpath, subimg)
                #copy(path_to_img, image_folder_path)
                #os.rename(os.path.join(image_folder_path,subimg), os.path.join(image_folder_path,str(img_num) + '.jpg'))
                
                #pad the new image with black pixels
                im = Image.open(os.path.join(subpath,subimg))
                new_im = Image.new("RGB", (1263, 1263))
                new_im.paste(im, (1263//2 - 500//2, 0))
                #resize image
                new_im = new_im.resize((img_size, img_size))
                new_im.save(os.path.join(image_folder_path,str(img_num) + '.jpg'))
                
                #figure out if there's a defect and update the csv
                annotation_img_name = subimg[:5] + '_label.bmp'
                annotation_img = Image.open(os.path.join(subpath,annotation_img_name))
                pix = list(annotation_img.getdata())
                if(sum(pix) > 0):
                    #this is an image with a defect. We will rotate this image 3 times to quadruple our positive dataset
                    writer.writerow({'image name': str(img_num) + '.jpg', 'defect status': 1})
                    print(1)
                    
                    #take last image, flip over y and save
                    content_img = Image.open(os.path.join(image_folder_path,str(img_num) + '.jpg'))
                    content_img = content_img.convert(mode = 'RGB')
                    img_num += 1
                    content_img = content_img.transpose(method=Image.FLIP_TOP_BOTTOM)
                    content_img.save(os.path.join(image_folder_path, str(img_num) + '.jpg'))
                    writer.writerow({'image name': str(img_num) + '.jpg', 'defect status': 1})
                    
                    #flip over x and save
                    content_img = Image.open(os.path.join(image_folder_path,str(img_num) + '.jpg'))
                    content_img = content_img.convert(mode = 'RGB')
                    img_num += 1
                    content_img = content_img.transpose(method=Image.FLIP_LEFT_RIGHT)
                    content_img.save(os.path.join(image_folder_path, str(img_num) + '.jpg'))
                    writer.writerow({'image name': str(img_num) + '.jpg', 'defect status': 1})
                    
                    #flip over y again and save
                    content_img = Image.open(os.path.join(image_folder_path,str(img_num) + '.jpg'))
                    content_img = content_img.convert(mode = 'RGB')
                    img_num += 1
                    content_img = content_img.transpose(method=Image.FLIP_TOP_BOTTOM)
                    content_img.save(os.path.join(image_folder_path, str(img_num) + '.jpg'))
                    writer.writerow({'image name': str(img_num) + '.jpg', 'defect status': 1})
                    
                else:
                    writer.writerow({'image name': str(img_num) + '.jpg', 'defect status': 0})
                    print(0)
```
